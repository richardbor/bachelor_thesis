### Imports

```python
import pandas as pd
import glob

import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.preprocessing import QuantileTransformer
```

### Load Data

```python
# load monthly data of mode based on selected months
def load_data(mode, months):
    first_read = True
    
    for month in months:
        data_month = pd.concat([pd.read_csv(file) for file in glob.glob('Data/Modes/data_supplemented_{}_{}.csv'.format(mode, month))], ignore_index = True)
        if (first_read):
            data_full = data_month
            first_read = False
        else:
            data_full = pd.concat([data_full, data_month], ignore_index=True)
    
    return data_full

# load full data of mode
def load_data_mode_full(mode):
    data_mode_full = pd.read_csv('Data/Modes/data_supplemented_full_{}.csv'.format(mode))
    
    return data_mode_full

# load POI data
def load_poi_data():
    poi_data = pd.read_csv('Data/POI_data.csv')
    
    return poi_data
```

```python
# define months for data loading (YYYYMM)
months = ['201911', '201912', '202001', '202002']

#load data by months
load_monthly_data = False

if (load_monthly_data):
    data_car = load_data('car', months)
    data_bicycle = load_data('bicycle', months)
    data_scooter = load_data('scooter', months)
    data_full = pd.concat([data_car, data_bicycle, data_scooter], ignore_index=True)
```

```python
# read in desired data
# set True if full data set load is desired
load_full_data = True

if (load_full_data):
    data_car = load_data_mode_full('car')
    data_bicycle = load_data_mode_full('bicycle')
    data_scooter = load_data_mode_full('scooter')
    data_full = pd.concat([data_car, data_bicycle, data_scooter], ignore_index=True)
```

### Inspection of data distribution and data transformation and scaling techniques

```python
data = data_full.copy()
# dropping features
data.drop(['id', 'provider', 'vehicleType', 'date_start', 'date_end', 'time_end', 'year', 'month', 'hour', 
           'longitude_start', 'longitude_end', 'latitude_start', 'latitude_end', 'coordinates_start', 'coordinates_end', 
           'distance', 'speed', 'time_basket', 'time_start',
           'weekday', 'weekend'], axis=1, inplace=True)
data.head()
```

```python
data['duration'] = data['duration'] / 60
```

```python
for column in data.columns:
    sns.displot(data, x=column)
```

#### Yeo-Johnson Transformation

```python
# Get column names
names = data.columns

data_transformed = data.copy()

pt = PowerTransformer(method='yeo-johnson', standardize=True)
data_transformed = pt.fit_transform(data_transformed)
data_transformed #= pd.DataFrame(data_transformed, columns=names)

data_transformed = pd.DataFrame(data_transformed, columns=names)
```

```python
data_transformed.hist()
plt.show()
```

```python
for column in data_transformed.columns:
    sns.displot(data_transformed, x=column)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=data_transformed.drop(columns=['POI_aero_start', 'POI_aero_end']))
plt.xticks(rotation=90)
```

```python
data_transformed.describe()
```

#### MinMax Scaling

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(data_transformed)

X_train_minmax = pd.DataFrame(X_train_minmax,columns=names) 
X_train_minmax.describe()
```

#### Normalization

```python
#Normalize
d_norm = data.copy()

# Get column names
names = d_norm.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit data on the scaler object
scaled_df = scaler.fit_transform(d_norm)
scaled_df = pd.DataFrame(scaled_df, columns=names)
  
# Normalizing the Data 
normalized_df = normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df,columns=names) 

for column in normalized_df.columns:
    sns.displot(normalized_df, x=column)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=normalized_df)
plt.xticks(rotation=90)
```

```python
normalized_df.describe()
```

#### Quantile Transformation

```python
d_quant = data.copy()
quantile_transformer = preprocessing.QuantileTransformer(output_distribution='normal', random_state=0)
d_trans = quantile_transformer.fit_transform(d_quant)
quantile_transformer.quantiles_

d_quant = pd.DataFrame(d_trans, columns=names)
```

```python
for column in d_quant.columns:
    sns.displot(d_quant, x=column)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=d_quant)
plt.xticks(rotation=90)
```

```python
d_quant.describe()
```
